## 🔷 5. Variables

> Variable = box jisme hum koi value store karte hain

### 📌 Rules:

* Start with letter or underscore (`_`)
* No space, no special symbol like `@`, `#`, etc.
* Case sensitive: `Name` ≠ `name`

```python
name = "Rohit"
age = 23
is_student = True
```

---

## 🔷 6. Data Types (Intro Level)

| Data Type | Example     |
| --------- | ----------- |
| int       | 1, 50, -10  |
| float     | 3.14, -0.5  |
| str       | "hello"     |
| bool      | True, False |

```python
print(type(23))          # int
print(type(3.14))        # float
print(type("Rohit"))     # str
print(type(True))        # bool
```

## 🔷 7. `input()` Function

> User se input lene ke liye

```python
name = input("Enter your name: ")
print("Hello,", name)
```

### 🔍 Note:

* Input value **hamesha string** type hoti hai.
* Convert to int or float like this:

```python
age = int(input("Enter your age: "))
pi = float(input("Enter PI value: "))
```

---

## 🔷 8. Type Conversion

```python
a = "10"
b = int(a)  # convert string to int

print(type(a))  # str
print(type(b))  # int
```
